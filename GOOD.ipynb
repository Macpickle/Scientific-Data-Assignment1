# Manipulation of Data


### The required tools

To develop this data into a clean data set some Python imports are needed to be able to break up the data into neat arrays to manipulate it into several ways. Pandas, MatplotLib and numpy are used to develop data further for analysis via these imports:

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

### Object Oriented Data collection

To begin, a class was defined to be able to parse and break up the data in a neat and organized fashion to make the program easier to understand.

When a class is defined, the constructor (**\_\_init\_\_**) is the first thing called, this step will initalize the class and provide class attribute values, then **loadfile** is called to break up the csv file provided.

```python
class DataCleanser:
    def __init__(self, file):
        self.file = file # set class attribute file to the file provided
        self.loadfile() # load the file 
```

### Cleaning and Manipulation of Data

In the **loadfile** class, the csv file is read using pandas with the **read_csv** function which will take the data and parse it into a pandas DataFrame. For this dataset, the only necessary columns are <ins>Day_of_Week</ins> and <ins>Time</ins>, so we will only use those from the file.

To begin, the <ins>Day of week</ins> column is first broken down for further comprehension. Panadas provides **value_counts** to count the number of unique values, then **sort_index** sorts the fields by index.

Then, A similar process is given to the <ins>Time</ins> column. To utilize the time given through the data, we must first convert it to a pandas **date-time** object to be able to manipulate it. It is then rounded to the nearest 30 minute interval, as using all times given provides a difficult to read visualization of the data.

Then, the data is broken down into its **value_counts** and sorted.

Following this, the data is further broken down into python dictionaries to be used for visualization and data collection,

Here is the code compiled into a single cell.

```python
class DataCleanser(DataCleanser):
    def loadfile(self):
        # reads csv then formats for specific columns
        self.data = pd.read_csv(self.file, usecols=["Day_of_week", "Time"])
        self.days = ["Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday"]
        
        # clean data, counts occurrences of days in week
        self.data_dayofweek = self.data["Day_of_week"].value_counts()
        self.data_dayofweek = self.data_dayofweek.reindex(self.days, fill_value=0)
        
        # convert time column to max within 30 minutes
        self.data["Time"] = pd.to_datetime(self.data["Time"], format="%H:%M:%S").dt.round("30min") # round to nearest 30 min interval
        self.data["Time"] = self.data["Time"].dt.strftime("%H:%M") # just get time from to_datetime
    
        # clean data, count occurrences of each time slot
        self.data_timeofday = self.data["Time"].value_counts()
        self.data_timeofday = self.data_timeofday.sort_index()
        
        # put into a dictionary
        self.dayofweek = {
            "x_axis": self.days,
            "x_axis_title": "Day of the Week (Day)",
            "y_axis": self.data_dayofweek.values,
            "y_axis_title": "Number of Accidents",
            "graph_title": "Number of Accidents per Day of the Week"
        }
        
        self.timeofday = {
            "x_axis": self.data_timeofday.index.astype(str),
            "x_axis_title": "Time of Day (Hour:Minutes)",
            "y_axis": self.data_timeofday.values,
            "y_axis_title": "Number of Accidents",
            "graph_title": "Number of Accidents per Time of Day"
        }
```

# Calculations

Following the collection and manipulation of data, the datasets are broken up into common set averages, calculations and values to illustrate more identities of the data. For this step, numpy is used by making use of it's **mean** and **median** functions that takes a pandas DataFrame, and calculate those values.
Following this, the line of best fit is calculated for this dataset.

```python
class DataCleanser(DataCleanser):
    def calculations(self):
        # calculate averages
        self.averages["day_mean"] = round(self.data_dayofweek.mean(),2)
        self.averages["day_median"] = round(self.data_dayofweek.median(), 2)
        
        self.averages["time_mean"] = round(self.data_timeofday.mean(), 2)
        self.averages["time_median"] = round(self.data_timeofday.median(), 2)
```

# Visualization

To visualize this data, Matplotlib is used to generate a bar and line chart to fully illustrate the findings with this data using the **graph** class method.
First, it plots the main points given by the dataset calculated above in the [Manipulation of Data](#Manipulation-of-Data) section. Extra styling is provided for a cleaner graph.

Then, the line of best fit is calculated with the provided dataset making use of numpys' **polyfit** function, which finds the closest function that represents the data set, and the **poly1d** function that defines a polynomial function.

Following this, extra styling is provided such as titles, grids and label directions is provided and then shown to the user.

Here is the code compiled into a single cell.

```python
class DataCleanser(DataCleanser):
    def graph(self, dataset):
        plt.plot(dataset["x_axis"], dataset["y_axis"])
        plt.fill_between(dataset["x_axis"], dataset["y_axis"], color="grey", alpha=0.4)
        
        # formulate line of best fit
        bestfit = np.polyfit(range(len(dataset["x_axis"])), dataset["y_axis"], 1)
        polynomial = np.poly1d(bestfit)
        plt.plot(dataset["x_axis"], polynomial(range(len(dataset["x_axis"]))))
        
        # styling and labels
        plt.xlabel(dataset["x_axis_title"])
        plt.xticks(rotation=90) 
        plt.ylabel(dataset["y_axis_title"])
        plt.title(dataset["graph_title"])
        plt.grid(axis = "y")
        
        # create window to show graph
        plt.show()
```
